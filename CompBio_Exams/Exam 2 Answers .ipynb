```python
import pymysql.cursors  
 
# Connect to the database.
connection = pymysql.connect(host='localhost',
                             user='root',
                             password='root',                             
                             db='exam2',
                             charset='utf8mb4',
                             cursorclass=pymysql.cursors.DictCursor)
 
print ("connect successful!!")
# okay, this worked! we are connected to mysql now

#file opening

germplasmacc = open("/home/osboxes/Course/Accelerated_Intro_WilkinsonExams/Germplasm.tsv", "r") # open the file with the data
germplasmacc.seek(0) # to start reading the file again, just in case
germplasmacc.readline() # to skip the first line

locusacc = open("/home/osboxes/Course/Accelerated_Intro_WilkinsonExams/LocusGene.tsv", "r") # open the file with the data
locusacc.seek(0)
locusacc.readline() # to skip the first line

import re # to use the regexp

try:
    # try will attemp to do what it says inside
    with connection.cursor() as cursor: # with: allows to safely exit the db if something goes wrong or in the end
        
        #loop for germplasm table:
        for line in germplasmacc.readlines(): # create the loop to read every line in the TSV, read lines
            line = line.rstrip()
            germplasmacc.readline()
            locus,germplasm,phenotype,pubmed = line.split('\t')# Read a single record
            pubmed = int(pubmed)
            sql = "insert into germplasm(locus,germplasm,phenotype,pubmed) values('%s', '%s', '%s', '%s')" % (locus,germplasm,phenotype,pubmed)
            cursor.execute(sql)                
        gp_id = 1    
        #loop for germplasm table:
        for line in locusacc.readlines(): # create the loop to read every line in the TSV, read lines

            line = line.rstrip()
            locusacc.readline()
            locus,gene,protein_length = line.split('\t')# Read a single record
            protein_length = int(protein_length)
           
            sql = "insert into locusgene(locus,gene,protein_length,gp_id) values('%s', '%s', '%s', '%s')" % (locus,gene,protein_length, gp_id)
            cursor.execute(sql)
            gp_id +=1
                
        germplasmacc.close()
        locusacc.close()
        
finally:
    print("")
    connection.commit()
    connection.close()
```
